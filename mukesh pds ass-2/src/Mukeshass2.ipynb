1)MODE:
The most detect to impute missing values using the mode when working with ordinal or categorical data, such as a dataset's seat count. By replacing missing values with the value that occurs the most frequently, it makes sure that no data is lost and aids in maintaining the integrity of the dataset.

```python
#seats mode
```

```python
seats_mode = ds['Seats'].mode()[0]
seats_mode
```

```python
ds['Seats'].fillna(seats_mode, inplace=True)
ds
```

```python
missing_values = ds.isnull().sum()
missing_values
```

2)MEDIAN:
The reason for using the median to impute missing values in the "Mileage" column is that mileage data usually has an uneven distribution. The median yields a representative value and is resistant to outliers when used as a measure of central tendency. It also guarantees consistency in analysis and maintains the integrity of the data.

```python
#mileage median
```

)I have Removed the units from some of the attributes and only kept the numerical values for kmpl from “Mileage”.

```python
ds['Mileage'] = ds['Mileage'].str.replace(r'kmpl', '', regex=True)
ds['Mileage'] = ds['Mileage'].str.replace(r'km/kg', '', regex=True)
```

```python
mileage_median = ds['Mileage'].median()
mileage_median
```

```python
ds['Mileage'].fillna(mileage_median, inplace=True)
ds
```

MEAN:
It makes sense to impute the mean to the missing values in the "Engine" column since it gives an indication of central tendency that corresponds to the average engine size in the dataset. By preserving the integrity of the dataset and guaranteeing that imputed values align with observed data, this method reduces the possibility of biases in later analyses.
